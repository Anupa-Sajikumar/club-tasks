## 01. Web Scrapping
### Can you write a Python script that scrapes (web-scraping) information about the 10 most COVID affected countries with the details (total cases, total deaths, and total recovered) from worldometers.info and write it to a CSV file.

Requesting the web page

```python
import os # for navigating the machine such as file directries
import requests # for requesting urls
import csv # handles csv files
import pandas as pd # handles data
from datetime import datetime # works with date and time
from bs4 import BeautifulSoup as soup # parses web pages
print("Necessary Modules Has Imported Sucessfully!!!")
date = datetime.now().strftime("%Y-%m-%d")
```

```python
date
```

```python
# Defining the URL where the page is accessible
url="https://www.worldometers.info/coronavirus/#countries"
# requesting the url
res = requests.get(url,allow_redirects=True) # res=responses
#headers
res.status_code
# checks if the page requested is done successfully or not
```

```python
res.headers
```

```python
res.text[:1000]
```

```python
res.text[:2000]
```

Parsing the web page

```python
soup_res = soup(res.text,"html.parser") # parses the text as a beautiful soup object
soup_samp = soup(res.text[:1000],"html.parser")#parses a sample of the text content
soup_samp
```

```python
sections = soup_res.find_all("div", id="maincounter-wrap")
sections
```

```python
len(sections)
```

```python
for section in sections:
    print("***************")
    print(section)
    print("***************")
    print("\r")
```

```python
cases = sections[0].find("span").text.replace(" ","").replace(",","")
deaths = sections[1].find("span").text.replace(",","")
recover = sections[2].find("span").text.replace(",","")
print("Number of cases: {}; Deaths: {}; and Recoveries: {}.".format(cases,deaths,recover))
```

Saving the results from the scrape

```python
#Creating the folder
try:
    os.mkdir("./downloads")
except:
    print("Folder exists already ")
    print("Unable to crewate the folder")
```

```python
# Writing the results in csv file
date=datetime.now().strftime("%d-%m-%Y") #Get the date format as DD-MM-YYYY
print(date)
```

```python
variables = ["Cases", "Deaths","Recoveries"]
outfile="./downloads/covid-19-statistics-"+date+".csv"
obs = cases,deaths,recover
print(obs)
```

```python
with open(outfile,"w",newline="") as file:
    writer = csv.writer(file)
    writer.writerow(variables)
    writer.writerow(obs)
```

```python
#Checking the presence of the file in downloads folder or not
os.listdir("./downloads")
# Opening the file and importing the contents of the file
with open(outfile,"r") as f:
    data= f.read()
print(data)
```

```python
table = soup_res.find("table",id="main_table_countries_today").find("tbody")
rows = table.find_all("tr",style="")
```

```python
#Extarct the information in each row in the table
table
rows
```

```python
global_info = []
for row in rows:
    columns = row.find_all("td")
    country_info = [column.text.strip() for column in columns]
    del country_info[7]
    global_info.append(country_info)
print(global_info[0:10])
print("\r")
print("Number of rows in tables: {}".format(len(global_info)))
print("\r")

del global_info[0] # delete first row containing world statistics
```

```python
try:
    os.mkdir("./downloads")
except:
    print("Already exists")

variables = ["Number","Country"," Total Cases", "New Cases","Total Deaths",
"New Deaths","Total Recovered", "Active Cases","Serious_Critical","Total Cases Per In Pop",
"Death Per In Pop","Total Tests","Tests Per In Pop","Population"]
outfile = "./downloads/covid-19-statistics-"+date+".csv"
print(outfile)
with open(outfile,"w",newline="") as f:
    writer = csv.writer(f)
    writer.writerow(variables)
    for country in global_info:
        writer.writerow(country)
```
